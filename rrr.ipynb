```python
def plot_those_circs(s,a,b,c,d):
    r1 = a
    r2 = b
    x_w = c
    y_w = d
    density = int(s)
    theta_d = 2*np.pi/density
    n_d = density
    x = np.zeros(density)
    y = np.zeros(density)
    for i in range(0,density):
        x[i] = r1*np.cos(theta_d*i)
        y[i] = r1*np.sin(theta_d*i)

    circum = int(s)
    xs = x
    ys = y
    for p in range(0,circum):
        xi = np.zeros(density)
        yi = np.zeros(density)
        for i in range(0,density):
            xi[i] = x_w*x[p]+r2*np.cos(theta_d*i)
            yi[i] = y_w*y[p]+r2*np.sin(theta_d*i)
        xs = np.append(xs,xi)
        ys = np.append(ys,yi)
    
    # title
    tit = ''
    ss = str(s)
    if (s % 2 == 0):
        tit = tit.join(['Even, s =',ss])
    else:
        tit = tit.join(['Odd, s =',ss])
    
    # overlapping dots? different color
    xs = np.array(xs)
    ys = np.array(ys)
    x_1 = []
    x_2 = []
    x_3 = []
    y_1 = []
    y_2 = []
    y_3 = []
    for bb in range(1,len(xs)):
        x_ss = xs[bb]
        index = np.where(xs == x_ss)
        y_pop = ys[index]
        for tt in range(1,len(index[0])):
            y_ss = ys[index]
            y_ind = np.where(y_pop == y_ss)
            if (len(y_ind[0])==1):
                x_1 = np.append(x_1,x_ss)
                y_1 = np.append(y_1,y_ss[0])
            elif (len(y_ind[0])==2):
                x_2 = np.append(x_2,x_ss)
                y_2 = np.append(y_2,y_ss[0])
                x_ss = np.delete(x_ss,index)
                y_ss = np.delete(y_ss,index)
            else:
                x_3 = np.append(x_3,x_ss)
                y_3 = np.append(y_3,y_ss)
                x_ss = np.delete(x_ss,index)
                y_ss = np.delete(y_ss,index)

    
    #plot with different colors
    plt.scatter(x_1,y_1, color = 'red')
    plt.scatter(x_2,y_2, color = 'blue')
    plt.scatter(x_3,y_3, color = 'green')
    plt.axis('equal')
    plt.title(tit)
    plt.show()

interact(plot_those_circs, 
         s = widgets.FloatSlider(value=12,
                                 min=1,
                                 max=34,
                                 step=1),
         a = widgets.FloatSlider(value=1.0,
                                 min=1.0,
                                 max=10.0,
                                 step=0.5),
         b = widgets.FloatSlider(value=8.0,
                                 min=1.0,
                                 max=10.0,
                                 step=0.5),
         c = widgets.FloatSlider(value=5.0,
                                 min=1.0,
                                 max=10.0,
                                 step=0.5),
         d = widgets.FloatSlider(value=5.0,
                                 min=1.0,
                                 max=10.0,
                                 step=0.5));
```
